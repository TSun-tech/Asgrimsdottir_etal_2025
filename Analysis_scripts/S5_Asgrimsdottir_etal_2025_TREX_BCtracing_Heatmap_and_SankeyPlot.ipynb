## visualisation script for generating complex heatmap of cell population marker gene expressions
#### script requires seperated exportation of gene expression and metadata matrix

```r
#for regulating plot sizes
pdim<-function(x,y){
    return(options(repr.plot.width=x, repr.plot.height=y))
}
```

```r
library(ComplexHeatmap)
library(circlize)
library(viridis)
library(magick)
```

```r
gene_expression<-read.csv(".../Output/Final_Emilia_EStracing_scRNA_expression_scaled.csv",
                         row.names = 1)

head(gene_expression)
```

```r
# marker of interest
genes <- c('SOX2', 'SOX6', 'HES1', 'FABP7', 'VIM', 
           'SOX9', 'SLC1A3', 'NTN1', 'SHH', 'WNT1', 'FGF17', 
           'TPBG', 'PTPRO', 'EMX2', 'PAX5', 'PAX6',  'HOXB2', 'OTX2', 
           'LMX1A', 'LMX1B', 'FOXA1', 'FOXA2', 'EN1', 'NEUROG1', 'NEUROG2', 'NEUROD1', 
           'ASCL1', 'NHLH1', 'NHLH2', 'DCX', 'NR4A2', 'PBX1', 'LMO3',  'PITX2', 
           'BARHL1', 'ERBB4', 'ISL1', 'TH', 'DDC', 
           'SLC18A1', 'SLC17A6',"NKX6-1","NKX6-2","NKX2-2",
           'POU4F1',  'GAD1'
           #,'ISL1','DLX1',"DBX1",'PITX3','HOXA2','FEV', 
          )
```

```r
gene_sub<-as.matrix(gene_expression[genes,])
gene_sub
```

```r
meta<-read.csv( ".../Output/Emilia_EStracing_scRNA_metadata.csv",
              row.names = 1)
```

```r
meta$Celltype_fct<-factor(meta$Celltype_fct,
                             levels = c( 'progBP', 'progFP',
                                              'NProg','Rgl1','Rgl3',
                                              'TransitionProg',
                                              'GliaProg',
                                        'NbML','OMTN','RN',
                                        'NbGaba','GABA',
                                        'NbGlut','Glut_neurons',
                                              'NbDA','DA_neuron',
                                        NA))

meta$diff_day<-factor(meta$diff_day, levels = c("d17", "d28"))
```

```r
# Create two sets of annotations
annotation_celltype <- meta$Celltype_fct
annotation_diffday <- meta$diff_day

# Combine annotations into a data frame
annotations_df <- data.frame(Cells = colnames(gene_sub), 
                              Annotation1 = annotation_celltype,
                              Annotation2 = annotation_diffday)

# Create a combined annotation for ordering
annotations_df$Combined <- interaction(annotations_df$Annotation1, annotations_df$Annotation2)

# Create a heatmap annotation object
ha_col <- HeatmapAnnotation(Celltype = annotations_df$Annotation1,
                            DiffDay = annotations_df$Annotation2,
                           col = list(Celltype = c("progBP" = "#fc8d62",
                                                   "progFP" = "#8da0cb",  
                                                   "NProg" = "#E69F00",
                                                   "Rgl1" = "#e78ac3",
                                                   "Rgl3" = "#a6d854",
                                                   "TransitionProg" = "#e5c494",
                                                   "GliaProg" = "#4daf4a",
                                                   "NbDA" = "#56B4E9",
                                                   "NbGlut" = "#BC8F8F",
                                                   "NbGaba" = "#a65628",
                                                   "NbML" = "#ff7f00",
                                                   "DA_neuron" = "#0072B2",
                                                   "RN" = "#e41a1c",
                                                   "OMTN" = "#66c2a5",
                                                   "GABA" = "#ffd92f",
                                                   "Glut_neurons" = "#984ea3"),
                                     DiffDay = c("d17" = "#D294AF", 
                                                 "d28" = "#79AEB2")
                                     )
                           )

# Define order of columns based on the combined annotation
ordered_cells <- annotations_df$Cells[order(annotations_df$Combined)]
```

```r
# Create the heatmap
ht <- Heatmap(gene_sub, 
               name = "Expression", 
               top_annotation = ha_col, 
               show_column_names = FALSE,
               cluster_columns = FALSE,
              cluster_column_slices = FALSE,
                show_row_names = TRUE,
               column_order = ordered_cells,
              column_split  = annotations_df$Annotation1,
             col = viridis(256),
              column_gap = unit(0, "mm"),
              use_raster = TRUE
             )

pdim(15, 9)
draw(ht)
```

## sankey plot

```r
# Libraries
library(tidyverse)
library(viridis)
library(patchwork)
library(hrbrthemes)
library(circlize)
library(networkD3)
```

```r
output_dir<-"...Output/"

df_chord<-read.csv(file = paste0(output_dir, 
                                 "Emilia_ESC_TREX_repBC_d17tod28_avgPercentage.csv"),
                  row.name =  "X")
```

```r
df_chord
```

```r
# I need a long format
data_long <- df_chord %>%
  rownames_to_column %>%
  gather(key = 'key', value = 'value', -rowname) %>%
  filter(value > 0)
colnames(data_long) <- c("source", "target", "value")
data_long$target <- paste(data_long$target, " ", sep="")

# From these flows we need to create a node data frame: it lists every entities involved in the flow
nodes <- data.frame(name=c(as.character(data_long$source), as.character(data_long$target)) %>% unique())

# With networkD3, connection must be provided using id, not using real name like in the links dataframe.. So we need to reformat it.
data_long$IDsource=match(data_long$source, nodes$name)-1
data_long$IDtarget=match(data_long$target, nodes$name)-1

# prepare colour scale
ColourScal ='d3.scaleOrdinal() .range(["#FDE725FF","#B4DE2CFF","#6DCD59FF","#35B779FF","#1F9E89FF","#26828EFF","#31688EFF","#3E4A89FF","#482878FF","#440154FF"])'

# Make the Network
sankeyNetwork(Links = data_long, Nodes = nodes,
                     Source = "IDsource", Target = "IDtarget",
                     Value = "value", NodeID = "name",
                     sinksRight=FALSE, colourScale=ColourScal, nodeWidth=40, fontSize=13, nodePadding=20)
```

```r
#now read in required matrix
meta<-read.csv(".../Output/Final_Emilia_EStracing_scRNA_metadata.csv", 
               row.names = 1)
```

```r
#subset to only have 394 cells that having cross timepoints barcode
meta<-subset(meta, subset = Clone_rep == "Rep")

dim(meta)
```

```r
head(meta)
```

```r
bc_count <- meta %>%
  group_by(Clone_ID, Celltype_refine, diff_day) %>%
  summarise(Count = n())
```

```r
# Filter for diff_day 17 and 28
df_day17 <- bc_count %>% filter(diff_day == "d17")
df_day28 <- bc_count %>% filter(diff_day == "d28")

# Merge the dataframes based on Clone_ID
merged_df <- merge(df_day17, df_day28, by = "Clone_ID", suffixes = c("_d17", "_d28"))

# Select relevant columns
result <- merged_df %>%
  select(Clone_ID, Celltype_refine_d17, Celltype_refine_d28, Count_d17, Count_d28)

head(result, 20)
```

```r
result_sum<- result %>% group_by(Celltype_refine_d17, Celltype_refine_d28) %>% summarise(Count = n())
```

```r
head(result_sum,20)
```

```r
colnames(result_sum)<-c("source", "target", "value")
```

```r
# From these flows we need to create a node data frame: it lists every entities involved in the flow
nodes <- data.frame(name=c(as.character(result_sum$source), 
                           as.character(result_sum$target)) %>% unique())

# With networkD3, connection must be provided using id, not using real name like in the links dataframe.. So we need to reformat it.
result_sum$IDsource=match(result_sum$source, nodes$name)-1
result_sum$IDtarget=match(result_sum$target, nodes$name)-1

# prepare colour scale
ColourScal ='d3.scaleOrdinal() .range(["#FDE725FF","#B4DE2CFF","#6DCD59FF","#35B779FF","#1F9E89FF","#26828EFF","#31688EFF","#3E4A89FF","#482878FF","#440154FF"])'

# Make the Network
(sankey<-sankeyNetwork(Links = result_sum, Nodes = nodes,
                     Source = "IDsource", Target = "IDtarget",
                     Value = "value", NodeID = "name",
                     sinksRight=FALSE, colourScale=ColourScal, nodeWidth=40, fontSize=13, nodePadding=20)
 )

print(sankey)
```
